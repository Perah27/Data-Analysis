```python
import numpy as np
```

```python
np.random.seed(100)
```

```python
angka = []
for i in range(0,100):
    angka.append(i)
```

```python
baru = pd.DataFrame(np.random.rand(100,5),angka,['A','B','C','D','E'],)
```

```python
baru.head()
```

```python
baru.plot.area(alpha=0.1) # menampilkan visualisasi seperti luas permukaan
```

```python
baru2 = pd.DataFrame(np.random.rand(4,5),[0,1,2,3],['A','B','C','D','E'],)
```

```python
baru2
```

```python
baru2.plot.bar()
```

```python
baru2.plot.bar(stacked=True)
```

```python
baru.plot.scatter(x='A', y='C', c='E') # sumbu x dari kolom A ke sumbu y kolom C, kolom E sebagai warnanya
```

```python
baru.plot.scatter(x='A', y='B', c='E', cmap='ocean')
```

```python
baru.plot.scatter(x='A', y='B', c='E', cmap='Reds')
```
